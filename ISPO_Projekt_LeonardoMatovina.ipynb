```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression

# ===================================
# 1) Odabir značajki i ciljne varijable
# ===================================

# Odabir numeričkih performansi vozača
X = data[['Pole_Positions', 'Race_Wins',
        'Podiums', 'Fastest_Laps', 'Points', 'Win_Rate', 'Podium_Rate',
        'FastLap_Rate', 'Points_Per_Entry']]

# Ciljna varijabla: je li vozač prvak
y = data['Champion'].astype(int)

# ===================================
# 2) Standardizacija značajki
# ===================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ===================================
# 3) Podjela na trening i test set
# ===================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)

# ===================================
# 4) Modeli strojnog učenja
# ===================================

models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Naive Bayes': GaussianNB(),
    'Neural Network': MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=2000, random_state=42),
    'Linear Regression': LinearRegression()
}

# ===================================
# 5) Izračun i prikaz ROC krivulja
# ===================================

plt.figure(figsize=(10, 8))

for name, model in models.items():
    model.fit(X_train, y_train)

    if name != 'Linear Regression':
        y_pred = model.predict_proba(X_test)[:, 1]
    else:
        y_pred = model.predict(X_test)

    fpr, tpr, _ = roc_curve(y_test, y_pred)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

# Diagonalna linija – random classifier
plt.plot([0, 1], [0, 1], 'k--')

plt.xlim([0, 1])
plt.ylim([0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve – Predicting if a Driver Will Become World Champion')
plt.legend(loc='lower right')

plt.show()
```

```python
X_train, X_test, y_train, y_test, train_idx, test_idx = train_test_split(
    X_scaled, y, data.index, test_size=0.3, random_state=42
)

random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest.fit(X_train, y_train)

y_test_pred_rf = random_forest.predict(X_test)
y_test_prob_rf = random_forest.predict_proba(X_test)[:, 1]
data.loc[test_idx, "Predicted_Champion_RF"] = y_test_pred_rf
data.loc[test_idx, "Champion_Probability_RF"] = y_test_prob_rf
top_test_predicted = data.loc[test_idx].sort_values(
    "Champion_Probability_RF", ascending=False
)[[
    "Driver", "Race_Wins", "Pole_Positions", "Podiums",
    "Points", "Champion", "Predicted_Champion_RF", "Champion_Probability_RF"
]]

print("\nTop 30 vozača po Random Forest modelu (samo TEST podaci):")
top_test_predicted.head(30)
```

```python
errors_test_rf = data.loc[test_idx][
    data.loc[test_idx, "Champion"] != data.loc[test_idx, "Predicted_Champion_RF"]
][[
    "Driver", "Race_Wins", "Pole_Positions", "Podiums",
    "Points", "Champion", "Predicted_Champion_RF", "Champion_Probability_RF"
]].sort_values("Champion_Probability_RF", ascending=False)

print("\nGreške modela Random Forest (samo TEST skup):")
errors_test_rf.head(20)
```

```python

neural_network = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)

neural_network.fit(X_train, y_train)

y_test_pred_rf = neural_network.predict(X_test)
y_test_prob_rf = neural_network.predict_proba(X_test)[:, 1]
data.loc[test_idx, "Predicted_Champion_NN"] = y_test_pred_rf
data.loc[test_idx, "Champion_Probability_NN"] = y_test_prob_rf
top_test_predicted = data.loc[test_idx].sort_values(
    "Champion_Probability_NN", ascending=False
)[[
    "Driver", "Race_Wins", "Pole_Positions", "Podiums",
    "Points", "Champion", "Predicted_Champion_NN", "Champion_Probability_NN"
]]

print("\nTop 30 vozača po Neural Network modelu (samo TEST podaci):")
top_test_predicted.head(30)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

y_test_pred = random_forest.predict(X_test)

accuracy = accuracy_score(y_test, y_test_pred)
precision = precision_score(y_test, y_test_pred)
recall = recall_score(y_test, y_test_pred)
f1 = f1_score(y_test, y_test_pred)

print("=== Random Forest - Evaluacija modela ===")
print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1-score  : {f1:.4f}")

# Matrica zabune
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_test_pred))
```
