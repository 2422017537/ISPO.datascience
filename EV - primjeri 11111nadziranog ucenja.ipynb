# Vanost matrice korelacije u EDA

Matrica korelacije je kljuan alat u istraivakoj analizi podataka (EDA) jer nam prua uvid u odnose izmeu razliitih varijabli u naem skupu podataka. Evo zato je ona posebno korisna za odabir stupaca za daljnju analizu:

1. **Identifikacija znaajnih veza**: Matrica korelacije nam omoguuje da brzo uoimo koje varijable su meusobno snano povezane. Visoka pozitivna ili negativna korelacija moe ukazivati na vane odnose koje vrijedi detaljnije istraiti.

2. **Izbjegavanje multikolinearnosti**: U kontekstu prediktivnog modeliranja, matrica korelacije nam pomae identificirati varijable koje su moda redundantne. Ako dvije varijable imaju vrlo visoku korelaciju, moda nije potrebno koristiti obje u modelu, to moe poboljati njegovu stabilnost i interpretabilnost.

3. **Usmjeravanje daljnje analize**: Korelacije koje uoimo mogu nas voditi prema zanimljivim hipotezama ili pitanjima za daljnje istraivanje. Na primjer, ako uoimo neoekivanu korelaciju izmeu dvije varijable, to moe potaknuti dublju analizu tog odnosa.

4. **Odabir relevantnih znaajki**: Za vizualizacije i statistike analize, esto elimo odabrati najrelevantnije varijable. Matrica korelacije nam pomae identificirati koje varijable su najjae povezane s naom ciljnom varijablom (u ovom sluaju, to bi mogla biti cijena ili popularnost elektrinih vozila).

5. **Otkrivanje skrivenih obrazaca**: Ponekad, matrica korelacije moe otkriti neoekivane veze izmeu varijabli koje nisu odmah oite iz sirovih podataka. Ovo moe voditi do novih uvida o podacima.

6. **Priprema za naprednije analize**: Razumijevanje korelacija meu varijablama je esto prvi korak prije primjene naprednijih tehnika kao to su faktorska analiza ili analiza glavnih komponenti (PCA).

U naem sluaju, matrica korelacije nam je pomogla identificirati da postoji znaajna pozitivna korelacija izmeu elektrinog dometa i osnovne cijene (MSRP), to sugerira da vozila s veim dometom openito imaju viu cijenu. Takoer, uoili smo slabu pozitivnu korelaciju izmeu godine modela i cijene, to moe ukazivati na blagi trend poveanja cijena s novijim modelima. Ovi uvidi su nam pomogli usmjeriti nau daljnju analizu na ove kljune varijable i njihove odnose.

```python
# Uvoz potrebnih biblioteka za strojno uenje
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# Priprema finalnog dataseta za modeliranje (uklanjamo redove s preostalim NaN vrijednostima)
# Fokusirat emo se na kljune stupce za nae modele
model_data = data_cleaned2[['Model Year', 'Electric Range', 'Electric Vehicle Type']].dropna()

print("--- DataFrame spreman za modeliranje ---")
print(model_data.head())
print("\n")
```

# =============================================================================

```python
print("--- MODEL 1: LINEARNA REGRESIJA ---")
```

#
# POSLOVNI PROBLEM: Moemo li predvidjeti ELEKTRINI DOMET (Electric Range)
# vozila na temelju GODINE MODELA (Model Year)?
#
# Ovo je REGRESIJSKI problem jer predviamo kontinuiranu numeriku vrijednost.

# 1. Priprema podataka za regresiju
# X su ulazne znaajke (ono to znamo), y je cilj (ono to predviamo)

```python
X_reg = model_data[['Model Year']]  # X mora biti 2D array (DataFrame)
y_reg = model_data['Electric Range']   # y je 1D array (Series)

# 2. Podjela podataka na skup za treniranje i testiranje (80% za treniranje, 20% za test)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# 3. Kreiranje i treniranje modela
reg_model = LinearRegression()  # Kreiramo instancu modela
reg_model.fit(X_train_reg, y_train_reg) # Treniramo model na podacima za treniranje

print("Linearna regresija je istrenirana.")

# 4. Predikcija na testnim podacima
predictions_reg = reg_model.predict(X_test_reg)

# 5. Evaluacija modela
mse = mean_squared_error(y_test_reg, predictions_reg)
r2 = r2_score(y_test_reg, predictions_reg)

print(f"\nMean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2) Score: {r2:.2f}")
print(f"Na model objanjava otprilike {r2*100:.0f}% varijance u elektrinom dometu.")

# 6. Vizualizacija rezultata
plt.figure(figsize=(10, 6))
plt.scatter(X_test_reg, y_test_reg, color='blue', alpha=0.5, label='Stvarni podaci')
plt.plot(X_test_reg, predictions_reg, color='red', linewidth=2, label='Regresijski pravac (Model)')
plt.title('Linearna Regresija: Godina Modela vs. Elektrini Domet')
plt.xlabel('Godina Modela')
plt.ylabel('Elektrini Domet (milje)')
plt.legend()
plt.grid(True)
plt.show()
```

#
# POSLOVNI PROBLEM: Moemo li predvidjeti TIP ELEKTRINOG VOZILA (BEV vs. PHEV)
# na temelju ELEKTRINOG DOMETA (Electric Range) i GODINE MODELA (Model Year)?
#
# Ovo je KLASIFIKACIJSKI problem jer predviamo jednu od dvije kategorije.

```python
# 1. Priprema podataka za klasifikaciju
X_clf = model_data[['Model Year', 'Electric Range']]
y_clf_text = model_data['Electric Vehicle Type']

le = LabelEncoder()
y_clf = le.fit_transform(y_clf_text)

# 2. Podjela podataka na skup za treniranje i testiranje
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42, stratify=y_clf)

# 3. Kreiranje i treniranje modela
clf_model = DecisionTreeClassifier(max_depth=3, random_state=42)
clf_model.fit(X_train_clf, y_train_clf)

print("Stablo odluivanja je istrenirano.")

# 4. Predikcija na testnim podacima
predictions_clf = clf_model.predict(X_test_clf)


# =============================================================================
# 5. EVALUACIJA MODELA - SVE NA JEDNOM MJESTU
# =============================================================================
print("\n--- Evaluacija modela: Stablo Odluivanja ---")

# A) Osnovne metrike
accuracy = accuracy_score(y_test_clf, predictions_clf)
print(f"\nTonost (Accuracy): {accuracy*100:.2f}%")

print("\nClassification Report:")
print(classification_report(y_test_clf, predictions_clf, target_names=le.classes_))

# B) Matrica konfuzije
cm_tree = confusion_matrix(y_test_clf, predictions_clf)
tn, fp, fn, tp = cm_tree.ravel()

print("\n--- Matrica Konfuzije (Stablo Odluivanja) ---")
cm_df = pd.DataFrame(cm_tree, index=le.classes_, columns=le.classes_)
print(cm_df)
print("\n--- Interpretacija Matrice Konfuzije ---")
print(f"Ispravne predikcije (dijagonala):")
print(f"  - True Negatives (TN): {tn} -> Model je tono predvidio da je vozilo 'BEV'.")
print(f"  - True Positives (TP): {tp} -> Model je tono predvidio da je vozilo 'PHEV'.")
print(f"\nPogrene predikcije (izvan dijagonale):")
print(f"  - False Positives (FP): {fp} -> Model je POGRENO predvidio 'PHEV', a vozilo je zapravo bilo 'BEV'. (Greka Tipa I)")
print(f"  - False Negatives (FN): {fn} -> Model je POGRENO predvidio 'BEV', a vozilo je zapravo bilo 'PHEV'. (Greka Tipa II)")

# C) Vizualizacija matrice konfuzije
ime_modela = 'Stablo Odluivanja'
plt.figure(figsize=(8, 6))
sns.heatmap(cm_tree, annot=True, fmt='d', cmap='Blues', 
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predviena klasa')
plt.ylabel('Stvarna klasa')
plt.title(f'Matrica Konfuzije - {ime_modela}')
plt.show()


# =============================================================================
# 6. VIZUALIZACIJA SAMOG MODELA (Stabla Odluivanja)
# =============================================================================
print("\n--- Vizualizacija nauenih pravila modela ---")
plt.figure(figsize=(20, 10))

feature_names_list = X_clf.columns.tolist()
class_names_list = le.classes_.tolist()

plot_tree(clf_model, 
          feature_names=feature_names_list,
          class_names=class_names_list,
          filled=True, 
          rounded=True,
          fontsize=10,
          precision=2)

plt.title("Vizualizacija Stabla Odluivanja (max_depth=3)")
plt.show()
```

```python
print("\n\n--- MODEL 3: LOGISTI REGRESIJA (KLASIFIKACIJA) ---")
#
# Objanjenje: "Iako joj je 'regresija' u imenu, Logistika Regresija je
# moan i jedan od najpopularnijih algoritama za KLASIFIKACIJU."

# 1. Uvoz modela
from sklearn.linear_model import LogisticRegression

# 2. Kreiranje i treniranje modela
log_reg_model = LogisticRegression(random_state=42)
log_reg_model.fit(X_train_clf, y_train_clf)

print("Logistika regresija je istrenirana.")

# 3. Predikcija na testnim podacima
predictions_log_reg = log_reg_model.predict(X_test_clf)


# =============================================================================
# 4. EVALUACIJA MODELA - LOGISTIKA REGRESIJA
# =============================================================================
print("\n--- Evaluacija modela: Logistika Regresija ---")

# A) Osnovne metrike
accuracy_log_reg = accuracy_score(y_test_clf, predictions_log_reg)
print(f"\nTonost (Accuracy): {accuracy_log_reg*100:.2f}%")

print("\nClassification Report:")
# Dodajemo zero_division=0 da izbjegnemo warning kada je precision/recall 0
print(classification_report(y_test_clf, predictions_log_reg, target_names=le.classes_, zero_division=0))

# B) Matrica konfuzije
cm_log_reg = confusion_matrix(y_test_clf, predictions_log_reg)
tn, fp, fn, tp = cm_log_reg.ravel()

print("\n--- Matrica Konfuzije (Logistika Regresija) ---")
cm_df = pd.DataFrame(cm_log_reg, index=le.classes_, columns=le.classes_)
print(cm_df)

print("\n--- Interpretacija Matrice Konfuzije ---")
print(f"Ispravne predikcije:")
print(f"  - True Negatives (TN): {tn} -> Model je tono predvidio da je vozilo 'BEV'.")
print(f"  - True Positives (TP): {tp} -> Model je tono predvidio da je vozilo 'PHEV'.")
print(f"\nPogrene predikcije:")
print(f"  - False Positives (FP): {fp} -> Model je POGRENO predvidio 'PHEV', a vozilo je zapravo bilo 'BEV'.")
print(f"  - False Negatives (FN): {fn} -> Model je POGRENO predvidio 'BEV', a vozilo je zapravo bilo 'PHEV'. <-- OVDJE JE PROBLEM!")


# C) Vizualizacija matrice konfuzije
ime_modela = 'Logistika Regresija'
plt.figure(figsize=(8, 6))
sns.heatmap(cm_log_reg, annot=True, fmt='d', cmap='Oranges', # Koristimo drugu boju za laku razliku
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predviena klasa')
plt.ylabel('Stvarna klasa')
plt.title(f'Matrica Konfuzije - {ime_modela}')
plt.show()
```

### 2. Zato je Logistika Regresija "omanula"?

Najvanija lekcija dananje vjebe krije se u rezultatima **Logistike Regresije**. Iako je ukupna tonost (Accuracy) bila skoro 80%, detaljna analiza `Classification Reporta` pokazala je da model **nijedan jedini PHEV (Plug-in Hybrid) automobil nije uspio prepoznati**.

Ovo se dogodilo zbog problema **neuravnoteenih klasa (imbalanced classes)**. Budui da u naem datasetu ima puno vie BEV (cca 80%) nego PHEV (cca 20%) vozila, najlaki put za model da postigne visoku tonost bio je da **uvijek predvia veinsku klasu (BEV)** i potpuno ignorira manjinsku.

### 3. Zato su Modeli temeljeni na Stablima bili bolji?

**Stablo Odluivanja** i **Random Forest** su po svojoj prirodi puno otporniji na ovaj problem. Oni donose odluke traei "pravila" koja najbolje razdvajaju podatke. Lako su pronali jednostavno pravilo (vjerojatno vezano uz `Electric Range`) koje efikasno odvaja BEV od PHEV vozila.
