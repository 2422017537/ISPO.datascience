# Vanost matrice korelacije u EDA

Matrica korelacije je kljuan alat u istraivakoj analizi podataka (EDA) jer nam prua uvid u odnose izmeu razliitih varijabli u naem skupu podataka. Evo zato je ona posebno korisna za odabir stupaca za daljnju analizu:

1. **Identifikacija znaajnih veza**: Matrica korelacije nam omoguuje da brzo uoimo koje varijable su meusobno snano povezane. Visoka pozitivna ili negativna korelacija moe ukazivati na vane odnose koje vrijedi detaljnije istraiti.

2. **Izbjegavanje multikolinearnosti**: U kontekstu prediktivnog modeliranja, matrica korelacije nam pomae identificirati varijable koje su moda redundantne. Ako dvije varijable imaju vrlo visoku korelaciju, moda nije potrebno koristiti obje u modelu, to moe poboljati njegovu stabilnost i interpretabilnost.

3. **Usmjeravanje daljnje analize**: Korelacije koje uoimo mogu nas voditi prema zanimljivim hipotezama ili pitanjima za daljnje istraivanje. Na primjer, ako uoimo neoekivanu korelaciju izmeu dvije varijable, to moe potaknuti dublju analizu tog odnosa.

4. **Odabir relevantnih znaajki**: Za vizualizacije i statistike analize, esto elimo odabrati najrelevantnije varijable. Matrica korelacije nam pomae identificirati koje varijable su najjae povezane s naom ciljnom varijablom (u ovom sluaju, to bi mogla biti cijena ili popularnost elektrinih vozila).

5. **Otkrivanje skrivenih obrazaca**: Ponekad, matrica korelacije moe otkriti neoekivane veze izmeu varijabli koje nisu odmah oite iz sirovih podataka. Ovo moe voditi do novih uvida o podacima.

6. **Priprema za naprednije analize**: Razumijevanje korelacija meu varijablama je esto prvi korak prije primjene naprednijih tehnika kao to su faktorska analiza ili analiza glavnih komponenti (PCA).

U naem sluaju, matrica korelacije nam je pomogla identificirati da postoji znaajna pozitivna korelacija izmeu elektrinog dometa i osnovne cijene (MSRP), to sugerira da vozila s veim dometom openito imaju viu cijenu. Takoer, uoili smo slabu pozitivnu korelaciju izmeu godine modela i cijene, to moe ukazivati na blagi trend poveanja cijena s novijim modelima. Ovi uvidi su nam pomogli usmjeriti nau daljnju analizu na ove kljune varijable i njihove odnose.

# ROC krivulja

ROC (Receiver Operating Characteristic) krivulja je grafiki prikaz koji se koristi za evaluaciju performansi klasifikacijskih modela. Ova krivulja prikazuje odnos izmeu stope stvarno pozitivnih (True Positive Rate - TPR) i stope lano pozitivnih (False Positive Rate - FPR) pri razliitim pragovima klasifikacije.

## Kljuni elementi ROC krivulje:

- **True Positive Rate (TPR)**: Takoer poznata kao osjetljivost ili odziv, predstavlja udio stvarno pozitivnih instanci koje su ispravno identificirane.
- **False Positive Rate (FPR)**: Predstavlja udio negativnih instanci koje su pogreno klasificirane kao pozitivne. 
- **AUC (Area Under the Curve)**: Povrina ispod ROC krivulje, koja kvantificira ukupnu performansu modela. AUC vrijednost od 1.0 predstavlja savren model, dok 0.5 predstavlja model koji nije bolji od nasuminog pogaanja.

## Interpretacija ROC krivulje:

- Krivulja blia gornjem lijevom kutu grafa indicira bolju performansu modela.
- Dijagonalna linija od (0,0) do (1,1) predstavlja nasumino pogaanje.
- Toke na krivulji predstavljaju razliite pragove klasifikacije.

## Znaaj u analizi elektrinih vozila:

U kontekstu klasifikacije elektrinih vozila, ROC krivulja moe pomoi u:

1. Usporedbi razliitih modela klasifikacije (npr. Random Forest vs. Logistika regresija).
2. Odabiru optimalnog praga za klasifikaciju vozila u razliite kategorije.
3. Procjeni koliko dobro model razlikuje izmeu razliitih tipova elektrinih vozila.

Ova analiza je posebno korisna za prodajne predstavnike jer prua uvid u pouzdanost modela pri klasifikaciji vozila, to moe biti kljuno u prilagodbi prodajnih strategija razliitim segmentima trita elektrinih vozila.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
# Uvoz potrebnih biblioteka za analizu podataka, vizualizaciju i strojno uenje

# Priprema podataka
X = data_cleaned2[['Model Year', 'Electric Range', 'Base MSRP']]
# Odabir znaajki za model
y = (data_cleaned2['Electric Vehicle Type'] == 'Battery Electric Vehicle (BEV)').astype(int)
# Kreiranje binarne ciljne varijable (1 za BEV, 0 za ostale)

# Standardizacija znaajki
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Standardizacija znaajki radi bolje performanse modela

# Podjela podataka
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
# Podjela podataka na trening i test set (70% trening, 30% test)

# Inicijalizacija modela
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Naive Bayes': GaussianNB(),
    'Neural Network': MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42),
    'Linear Regression': LinearRegression()
}
# Kreiranje rjenika s razliitim modelima strojnog uenja

# Treniranje modela i raunanje ROC krivulja
plt.figure(figsize=(10, 8))
# Kreiranje nove figure za grafikon

for name, model in models.items():
    model.fit(X_train, y_train)
    # Treniranje modela na trening podacima
    if name != 'Linear Regression':
        y_pred = model.predict_proba(X_test)[:, 1]
        # Za klasifikacijske modele, koristimo vjerojatnosti pozitivne klase
    else:
        y_pred = model.predict(X_test)
        # Za linearnu regresiju, koristimo predviene vrijednosti
    fpr, tpr, _ = roc_curve(y_test, y_pred)
    # Izraun stope lano pozitivnih (FPR) i stvarno pozitivnih (TPR)
    roc_auc = auc(fpr, tpr)
    # Izraun podruja ispod ROC krivulje (AUC)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')
    # Crtanje ROC krivulje za svaki model

plt.plot([0, 1], [0, 1], 'k--')
# Dodavanje dijagonalne linije koja predstavlja sluajni klasifikator
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
# Postavljanje granica osi
plt.xlabel('Stopa lano pozitivnih')
plt.ylabel('Stopa stvarno pozitivnih')
plt.title('ROC krivulje za razliite modele')
plt.legend(loc="lower right")
# Dodavanje oznaka osi, naslova i legende
plt.show()
# Prikaz grafikona
```
