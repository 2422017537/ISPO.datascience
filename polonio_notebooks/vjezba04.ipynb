# Vanost matrice korelacije u EDA

Matrica korelacije je kljuan alat u istraivakoj analizi podataka (EDA) jer nam prua uvid u odnose izmeu razliitih varijabli u naem skupu podataka. Evo zato je ona posebno korisna za odabir stupaca za daljnju analizu:

1. **Identifikacija znaajnih veza**: Matrica korelacije nam omoguuje da brzo uoimo koje varijable su meusobno snano povezane. Visoka pozitivna ili negativna korelacija moe ukazivati na vane odnose koje vrijedi detaljnije istraiti.

2. **Izbjegavanje multikolinearnosti**: U kontekstu prediktivnog modeliranja, matrica korelacije nam pomae identificirati varijable koje su moda redundantne. Ako dvije varijable imaju vrlo visoku korelaciju, moda nije potrebno koristiti obje u modelu, to moe poboljati njegovu stabilnost i interpretabilnost.

3. **Usmjeravanje daljnje analize**: Korelacije koje uoimo mogu nas voditi prema zanimljivim hipotezama ili pitanjima za daljnje istraivanje. Na primjer, ako uoimo neoekivanu korelaciju izmeu dvije varijable, to moe potaknuti dublju analizu tog odnosa.

4. **Odabir relevantnih znaajki**: Za vizualizacije i statistike analize, esto elimo odabrati najrelevantnije varijable. Matrica korelacije nam pomae identificirati koje varijable su najjae povezane s naom ciljnom varijablom (u ovom sluaju, to bi mogla biti cijena ili popularnost elektrinih vozila).

5. **Otkrivanje skrivenih obrazaca**: Ponekad, matrica korelacije moe otkriti neoekivane veze izmeu varijabli koje nisu odmah oite iz sirovih podataka. Ovo moe voditi do novih uvida o podacima.

6. **Priprema za naprednije analize**: Razumijevanje korelacija meu varijablama je esto prvi korak prije primjene naprednijih tehnika kao to su faktorska analiza ili analiza glavnih komponenti (PCA).

U naem sluaju, matrica korelacije nam je pomogla identificirati da postoji znaajna pozitivna korelacija izmeu elektrinog dometa i osnovne cijene (MSRP), to sugerira da vozila s veim dometom openito imaju viu cijenu. Takoer, uoili smo slabu pozitivnu korelaciju izmeu godine modela i cijene, to moe ukazivati na blagi trend poveanja cijena s novijim modelima. Ovi uvidi su nam pomogli usmjeriti nau daljnju analizu na ove kljune varijable i njihove odnose.

```python
print("\n\n--- MODEL 3: K-MEANS KLASTERIRANJE ---")

# Uvoz potrebnih biblioteka
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 1. Priprema podataka za klasteriranje
# POSLOVNI PROBLEM: Moemo li pronai prirodne grupe (klastere) elektrinih vozila
# na temelju Godine Modela, Elektrinog Dometa i Cijene?

# Odabiremo numerike znaajke koje imaju smisla za grupiranje
cluster_features = ['Model Year', 'Electric Range', 'Base MSRP']
# Uklanjamo redove gdje bilo koja od ovih vrijednosti nedostaje
cluster_data = data_cleaned2[cluster_features].dropna()

# VAAN KORAK: Skaliranje podataka
# K-Means je osjetljiv na skalu podataka, pa moramo sve znaajke dovesti na istu skalu.
scaler = StandardScaler()
cluster_data_scaled = scaler.fit_transform(cluster_data)
print("Podaci za klasteriranje su pripremljeni i skalirani.")

# 2. Odabir optimalnog broja klastera 'K' (Elbow metoda)
print("\n--- Primjena Elbow metode za pronalazak optimalnog K ---")
wcss = [] # Within-Cluster Sum of Squares
k_range = range(1, 11)
for i in k_range:
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init='auto')
    kmeans.fit(cluster_data_scaled)
    wcss.append(kmeans.inertia_)

# Vizualizacija Elbow metode
plt.figure(figsize=(10, 6))
plt.plot(k_range, wcss, marker='o', linestyle='--')
plt.title('Elbow Metoda za Odabir Optimalnog K')
plt.xlabel('Broj klastera (K)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
# Objanjenje: "Grafikon pokazuje 'lakat' oko K=4, to sugerira da je 4 dobar broj klastera."
```

```python
# 3. Kreiranje i treniranje K-Means modela s odabranim K
optimal_k = 4
kmeans_model = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42, n_init='auto')
cluster_labels = kmeans_model.fit_predict(cluster_data_scaled)

# 4. Analiza i vizualizacija rezultata
# Dodajemo oznake klastera natrag u na originalni (neskalirani) DataFrame za laku interpretaciju
cluster_data['Cluster'] = cluster_labels

print(f"\n--- Rezultati klasteriranja s K={optimal_k} ---")

# Vizualizacija klastera
plt.figure(figsize=(12, 8))
sns.scatterplot(data=cluster_data, x='Electric Range', y='Base MSRP', hue='Cluster', palette='viridis', alpha=0.8, s=50)
plt.title(f'Prikaz Klastera Vozila (K={optimal_k})')
plt.xlabel('Elektrini Domet (milje)')
plt.ylabel('Cijena (Base MSRP)')
plt.legend(title='Klaster')
plt.grid(True)
plt.show()

# Prikaz prosjenih vrijednosti za svaki klaster
print("\n--- Prosjene karakteristike svakog klastera ---")
print(cluster_data.groupby('Cluster').mean().round(2))
```

```python
# =============================================================================
# MODEL 6: ASOCIJACIJSKA PRAVILA (Market Basket Analysis)
# =============================================================================
print("\n\n--- MODEL 6: ASOCIJACIJSKA PRAVILA ---")

# Osiguravamo da je biblioteka 'mlxtend' instalirana
!pip install --user -q mlxtend

from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# 1. Priprema podataka za asocijacije
# Fokusirat cemo se samo na TESLA vozila radi jednostavnosti
basket_data = data_cleaned2[data_cleaned2['Make'] == 'TESLA']
transactions = basket_data.groupby('Model Year')['Model'].apply(list).values.tolist()

# Transformacija podataka u format za apriori algoritam
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
basket_df = pd.DataFrame(te_ary, columns=te.columns_)

# 2. Primjena Apriori algoritma
frequent_itemsets = apriori(basket_df,  min_support=0.01, use_colnames=True)

# 3. Generiranje asocijacijskih pravila
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)

# 4. Prikaz rezultata
print("\n--- Najee kombinacije modela (Tesla) ---")
print(frequent_itemsets.sort_values(by='support', ascending=False).head())
print("\n--- Otkrivena asocijacijacijska pravila (Tesla) ---")
print(rules.sort_values(by='lift', ascending=False).head())

print("\n--- KRAJ VJEBE (NENADZIRANO UENJE) ---")
```
